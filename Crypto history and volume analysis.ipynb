```python
import pandas as pd
import time
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
import numpy as np
from pprint import pprint
import plotly.offline as py
import plotly.graph_objs as go
import plotly.figure_factory as ff
py.init_notebook_mode(connected=True)
```

```python
# get market info for bitcoin from the start of 2016 to the current day
bitcoin_market_info = pd.read_html("https://coinmarketcap.com/currencies/bitcoin/historical-data/?start=20160101&end="+time.strftime("%Y%m%d"))[0]
# convert the date string to the correct date format
bitcoin_market_info = bitcoin_market_info.assign(Date=pd.to_datetime(bitcoin_market_info['Date']))
# when Volume is equal to '-' convert it to 0
bitcoin_market_info.loc[bitcoin_market_info['Volume']=="-",'Volume']=0
# convert to int
bitcoin_market_info['Volume'] = bitcoin_market_info['Volume'].astype('int64')

bitcoin_market_info.head()
```

```python
btc_trace = go.Scatter(x=bitcoin_market_info['Date'], y=bitcoin_market_info['Close**'])
py.iplot([btc_trace])
```

```python
bitcoin_market_info.mean()
```

```python
btc_vol = go.Scatter(x=bitcoin_market_info['Date'], y=bitcoin_market_info['Volume'])
py.iplot([btc_vol])
```

```python
# get market info for ethereum from the start of 2016 to the current day
ethereum_market_info = pd.read_html("https://coinmarketcap.com/currencies/ethereum/historical-data/?start=20160101&end="+time.strftime("%Y%m%d"))[0]
# convert the date string to the correct date format
ethereum_market_info = ethereum_market_info.assign(Date=pd.to_datetime(ethereum_market_info['Date']))
# when Volume is equal to '-' convert it to 0
ethereum_market_info.loc[ethereum_market_info['Volume']=="-",'Volume']=0
# convert to int
ethereum_market_info['Volume'] = ethereum_market_info['Volume'].astype('int64')

ethereum_market_info.head()
```

```python
ethereum_market_info.mean()
```

```python
# get market info for ethereum from the start of 2016 to the current day
litecoin_market_info = pd.read_html("https://coinmarketcap.com/currencies/litecoin/historical-data/?start=20160101&end="+time.strftime("%Y%m%d"))[0]
# convert the date string to the correct date format
litecoin_market_info = litecoin_market_info.assign(Date=pd.to_datetime(litecoin_market_info['Date']))
# when Volume is equal to '-' convert it to 0
litecoin_market_info.loc[litecoin_market_info['Volume']=="-",'Volume']=0
# convert to int
litecoin_market_info['Volume'] = litecoin_market_info['Volume'].astype('int64')

litecoin_market_info.head()
```

```python
litecoin_market_info.mean()
```

```python
combined_data = pd.merge(bitcoin_market_info, ethereum_market_info, on="Date")
combined_data.columns = ["Date", "Bitcoin Open", "Bitcoin High", "Bitcoin Low", "Bitcoin Close", "Bitcoin Volume", "Bitcoin Market Cap",
                        "Ethereum Open", "Ethereum High", "Ethereum Low", "Ethereum Close", "Ethereum Volume", "Ethereum Market Cap"]
combined_data.head()
```

```python
combined_data2 = pd.merge(combined_data, litecoin_market_info, on='Date')
combined_data2.head()
combined_data2.columns = ["Date", "Bitcoin Open", "Bitcoin High", "Bitcoin Low", "Bitcoin Close", "Bitcoin Volume", "Bitcoin Market Cap",
                        "Ethereum Open", "Ethereum High", "Ethereum Low", "Ethereum Close", "Ethereum Volume", "Ethereum Market Cap",
                         "Litecoin Open", "Litecoin High", "Litecoin Low", "Litecoin Close", "Litecoin Volume", "Litecoin Market Cap"]
combined_data2.head()
```

```python
trace_1 = go.Scatter(x=combined_data['Date'], y=combined_data['Bitcoin Close'], name = 'Bitcoin')
trace_2 = go.Scatter(x=combined_data['Date'], y=combined_data['Ethereum Close'], name = 'Ethereum')
trace_3 = go.Scatter(x=combined_data2['Date'], y=combined_data2['Litecoin Close'], name = 'Litecoin')

data_1 = [trace_1, trace_2, trace_3]
layout=go.Layout(title="Closing Value Comparison", yaxis={'title':'Closing price(US$)'})
figure_1=go.Figure(data=data_1,layout=layout)
py.iplot(figure_1)
```

```python
vol_1 = go.Scatter(x=combined_data['Date'], y=combined_data['Bitcoin Volume'], name = 'Bitcoin')
vol_2 = go.Scatter(x=combined_data['Date'], y=combined_data['Ethereum Volume'], name = 'Ethereum')
vol_3 = go.Scatter(x=combined_data2['Date'], y=combined_data2['Litecoin Volume'], name = 'Litecoin')

data_2 = go.Data([vol_1, vol_2, vol_3])
layout=go.Layout(title="Volume Comparison", yaxis={'title':'Volume(US$)'})
figure=go.Figure(data=data_2,layout=layout)
py.iplot(figure)
```
